Data Uploading and Processing


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

calories=pd.read_csv("/content/calories.csv")
calories.head()
```

```python
exercise=pd.read_csv("/content/exercise.csv")
exercise.head()
```

```python
calories_data=pd.concat([exercise,calories['Calories']],axis=1)
calories_data.head()
```

```python
calories_data.shape
```

```python
calories_data.info()
```

```python
# Checking missing values

calories_data.isnull().sum()
```

Data Analysis & Visualization


```python
calories_data.describe()
```

```python
import seaborn as sns

sns.set()
sns.countplot(calories_data['Gender'])
```

```python
sns.distplot(calories_data['Age'])
```

```python
sns.distplot(calories_data['Height'])
```

```python
sns.distplot(calories_data['Weight'])
```

```python
# converting categorical data into numerical data

calories_data.replace({'Gender':{'male':0,'female':1}},inplace=True)
calories_data.head()
```

```python
# dropping user_id column

calories_data=calories_data.drop('User_ID',axis=1)
calories_data.head()
```

```python
# Correlation between Dataset

cor=calories_data.corr()
plt.figure(figsize=(10,6))
sns.heatmap(cor,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Reds')
```

```python
# separating features and target columns

X=calories_data.drop(columns=['Calories'],axis=1)
Y=calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

Model Training

```python
# XGBoost Regressor

from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb.fit(X_train,Y_train)
```

Model Evaluation

```python
#prediction on test data

test_data_pred=xgb.predict(X_test)
print(test_data_pred)
```

```python
Y_test.head()
```

```python
# Mean Absolute Error(MAE)
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(Y_test,test_data_pred)
print(MAE)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(Y_test,test_data_pred)
print(r2)
```

```python
r2=r2_score(Y_train,xgb.predict(X_train))
print(r2)
```

Evaluation for Random Forest Model

```python
from sklearn.ensemble  import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train,Y_train)
```

```python
test_data_pred=rf.predict(X_test)
print(test_data_pred)
```

```python
Y_test.head()
```
