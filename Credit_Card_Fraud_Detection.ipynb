```python
import pandas as pd
import numpy as np

df=pd.read_csv("/content/creditcard.csv")
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
#checking missing value in each column

df.isnull().sum()
```

```python
df['Class'].value_counts()
```

```python
#separating data for analysis

legit=df[df.Class==0]
fraud=df[df.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe()

```

```python
fraud.Amount.describe()
```

```python
#Compare values of transactions

df.groupby('Class').mean()
```

```python
#Under Sampling
#Build a sample dataset containing similar distribution of normal and legit transactions -> 492

legit_sample=legit.sample(n=492)
```

```python
#Concatenating two DataFrames

new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting data into features and target

X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
print(X)
```

```python
print(Y)
```

```python
#Split dataset into train and test sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
```

```python
# Model Training
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
# Evaluate model based on Accuracy score

from sklearn.metrics import accuracy_score
# Accuracy score on train dataset
X_train_pred=lr.predict(X_train)
train_data_accu=accuracy_score(X_train_pred,Y_train)
print('Accuracy score on train dataset : ',train_data_accu)
```

```python
# Accuracy score on test dataset

X_test_pred=lr.predict(X_test)
test_data_accu=accuracy_score(X_test_pred,Y_test)
print('Accuracy score on test dataset :',test_data_accu)
```

```python
from sklearn.metrics import classification_report, f1_score
# Make predictions on the test set
y_pred = lr.predict(X_test)

# Generate the classification report
print("Classification Report:\n", classification_report(Y_test, y_pred))

# Calculate and print the F1 score (macro-average)
f1 = f1_score(Y_test, y_pred, average='macro')
print("\nF1 Score (macro-average):", f1)
```

```python
# Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X_train,Y_train)
```

```python
# Evaluate model based on Accuracy score : RF Classifier Model
from sklearn.metrics import classification_report, f1_score
# Make predictions on the test set
y_pred = rf.predict(X_test)

# Generate the classification report
print("Classification Report:\n", classification_report(Y_test, y_pred))

# Calculate and print the F1 score (macro-average)
f1 = f1_score(Y_test, y_pred, average='macro')
print("\nF1 Score (macro-average):", f1)
```

```python
# Evaluate model based on Accuracy score : XGBoost Classification Model

from xgboost import XGBClassifier
from sklearn.metrics import classification_report, f1_score

xg=XGBClassifier(n_estimators=100,random_state=42)
xg.fit(X_train,Y_train)
```

```python
# Make predictions on the test set
y_pred = xg.predict(X_test)

# Generate the classification report
print("Classification Report:\n", classification_report(Y_test, y_pred))

# Calculate and print the F1 score (macro-average)
f1 = f1_score(Y_test, y_pred, average='macro')
print("\nF1 Score (macro-average):", f1)
```
